```python
import requests
import re
from openpyxl import load_workbook
import openpyxl
workbook = load_workbook(filename='Data.xlsx')
sheet1 = workbook['Sheet1']
print(sheet1.dimensions)
wb=openpyxl.Workbook() 
ws = wb.create_sheet(index=0,title='Sheet1') 
```

```python

#开始分问题
List = ['汪文斌', '耿爽', '华春莹','陆慷','赵立坚']
j = 1
for i in range(1,990):
    if sheet1.cell(row=i,column=8).value == 1:
        nameStr = sheet1.cell(row=i,column=2).value
        print(nameStr)
        if nameStr == '2021年7月29日外交部发言人赵立坚主持例行记者会':
            continue
        if sheet1.cell(row=i,column=9).value == 0:
            continue
        temp = sheet1.cell(row=i,column=6).value
        for val in temp.split('问：'):
            qList = val.split('答：')
            if(len(qList) >= 2):
                ws.cell(row = j, column=1).value = nameStr
                ws.cell(row = j, column=3).value = qList[0]
                ws.cell(row = j, column=4).value = qList[1]
                j += 1
            else:
                print(qList)  
string = r'第二类.xlsx'
wb.save(string)                 
                
                
                
                
```
