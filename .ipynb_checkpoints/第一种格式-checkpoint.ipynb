```python
import requests
import re
from openpyxl import load_workbook
workbook = load_workbook(filename='Data.xlsx')
sheet1 = workbook['Sheet1']
print(sheet1.dimensions)
```

```python
#这一段是把每天的记者会的描述类型找到，看看是'记者：',还是'问：'这两种形式的哪一种
times = 0
for i in range(1,990):
    nameStr = sheet1.cell(row=i,column=2).value
    temp = sheet1.cell(row=i,column=6).value
    count = temp.count('答：')
    USA =  temp.count('美国')+temp.count('中美')+temp.count('美方')
   
    if USA > 0:
        times += 1
        sheet1.cell(row=i,column=9).value = 1
    else:
        sheet1.cell(row=i,column=9).value = 0
    if count >= 3:
        sheet1.cell(row=i,column=8).value = 1
    else:
        count2 = temp.count('记者：')
        sheet1.cell(row=i,column=8).value = 0
print(times)
workbook.save('Data (2).xlsx')
    
```

```python
#开始分问题
wb=openpyxl.Workbook() 
ws = wb.create_sheet(index=0,title='Sheet1') 
List = ['汪文斌', '耿爽', '华春莹','陆慷','赵立坚']
j = 1
for i in range(1,990):
    if sheet1.cell(row=i,column=8).value == 0:
        nameStr = sheet1.cell(row=i,column=2).value
        print(nameStr)
        if sheet1.cell(row=i,column=9).value == 0:
            continue
        if nameStr == '2021年7月29日外交部发言人赵立坚主持例行记者会':
            continue

        name = ''
        for val in List:
            if val in nameStr:
                name = val
        temp = sheet1.cell(row=i,column=6).value
        sub = ''
        for val in re.findall(r'\n(.*?)记者：',temp):
            sub = sub + '((?:.|\n)*)'+ '\n' + val + '记者'
        t = 0

        for questions in re.findall(sub,temp)[0]:
            qList  = questions.split(name)
            if(len(qList) >= 2):
                    ws.cell(row = j, column=1).value = nameStr
                    ws.cell(row = j, column=2).value = name
                    ws.cell(row = j, column=3).value = qList[0].replace('：','')
                    ws.cell(row = j, column=4).value = qList[1].replace('：','')
                    ws.cell(row = j, column=5).value = re.findall(r'\n(.*?)记者',temp)[t]
                    j += 1
                    t += 1
            else:
                print(qList)

string = r'第一类.xlsx'
wb.save(string)  
```
