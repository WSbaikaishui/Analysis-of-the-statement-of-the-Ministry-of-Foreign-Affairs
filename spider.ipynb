```python
import requests
from bs4 import BeautifulSoup
import openpyxl 
from w3lib import html

from lxml import etree
```

```python
wb=openpyxl.Workbook() 
ws = wb.create_sheet(index=0,title='Sheet1') 
```

```python
count = 0;
for i in range(66):
    if i == 0:
        url = 'https://www.fmprc.gov.cn/web/fyrbt_673021/jzhsl_673025/default.shtml'
        
    else:
        url = 'https://www.fmprc.gov.cn/web/fyrbt_673021/jzhsl_673025/default_' + str(i) + '.shtml'
    res = requests.get(url)
    res.encoding = 'UTF-8'
    soup = BeautifulSoup(res.text)
    a_htmls = soup.find_all('div', class_="rebox_news")
    a_html = a_htmls[0].find_all('a')
    for a in a_html:
        count += 1
        ws.cell(row=count, column=1).value =a['href']
        ws.cell(row=count, column=2).value =a.contents[0]
#         print(a['href'])
#         print(a.contents[0])
count
wb.save('Data.xlsx')
```
